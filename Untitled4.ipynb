```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy
import tensorflow as tf
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# seed 값 생성
numpy.random.seed(3)
tf.random.set_seed(3)

#데이터 로드
dataset = numpy.loadtxt("modu_csv-master/pima-indians-diabetes.csv", delimiter = ",")
x = dataset[:,0:8]
y = dataset[:,8]

#모델의 설정
model = Sequential()
model.add(Dense(12, input_dim = 8, activation = 'relu'))
model.add(Dense(8, activation = 'relu'))
model.add(Dense(1, activation = 'sigmoid'))

#모델 컴파일
model.compile(loss = 'binary_crossentropy',
             optimizer = 'adam',
             metrics = ['accuracy'])

#모델 실행
model.fit(x, y, epochs = 200, batch_size = 10, verbose = False)
model.fit(x, y, epochs = 5, batch_size = 10)

#결과 출력
print('\n Accuracy: %.4f' %(model.evaluate(x,y)[1]))
```

```python
plt.figure(figsize = (12, 12)) # 만약 똑같은 2개의 결과가 나온다면 이것을 지우면 됨.

# heatmap() 함수는 두 항목씩 짝을 지어 각각 어떤 패턴으로 변화하는지 관찰하는 함수
# 두 항목이 전혀 다른 패턴으로 변화하고 있으면 0.
# 서로 비슷한 패턴으로 변화하고 있으면 1에 가까운 값을 출력한다.
# heatmap() 함수를 통해 각 항목 값의 상관관계를 나타낼 수 있다.

# vmax: 색상의 밝기 조절
# cmap: 미리 정해진 matplotlib 색상의 설정값 가져오기

sns.heatmap(df.corr(), linewidths = 0.1, vmax = 0.5, cmap = plt.cm.gist_heat, linecolor = 'white', annot = True)
plt.show()
```
